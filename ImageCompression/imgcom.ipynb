```python
from matplotlib.image import imread
import numpy as np
import matplotlib.pyplot as plt

A = imread('afghan.jpg')
B = np.mean(A, -1)

print('Image dimensions:', A.shape)

plt.figure()
plt.imshow(B, cmap=plt.get_cmap('gray'))
plt.axis('off')
```

```python
Bt = np.fft.fft2(B)
Btsort = np.sort(np.abs(Bt.reshape(-1)))

keep = 0.1
thresh = Btsort[int(np.floor((1-keep)*len(Btsort)))]
ind = np.abs(Bt)>thresh
Btlow = Bt * ind
Alow = np.fft.ifft2(Btlow).real

plt.figure()
plt.imshow(Alow, cmap=plt.get_cmap('gray'))
plt.axis('off')
```
